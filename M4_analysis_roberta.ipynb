```python
import pandas as pd
from tqdm import tqdm

train_df = pd.read_json('embeddingtest.json') # 根据给的训练数据提取完实体以及抽取具体问题之后的json
train_df
```

```python
train_df['target'] = 0
train_df
```

```python
for idx,item in tqdm(train_df.iterrows()):
    # print(item)
    try:
        if item['entity']==None:
            train_df['target'][idx] = 1
            continue
        elif item['query'][0]=='[' and item['query'][-1]==']':
            src_list = eval(item['query'])
        elif item['query'][0]=='[' and item['query'][-1]=='。':
            src_list = eval(item['query'][:-1])
        entity = item['entity']
        target_list = []
        for src_string in src_list:
            if entity == src_string:
                continue
            elif entity in src_string:
                src_string = src_string.replace('的','')
                target_string = src_string.replace(entity,'')
            else:
                target_string = src_string
            target_list.append(target_string)
        if len(target_list)==0:
            train_df['target'][idx]=2
        else:
            train_df['target'][idx]=str(target_list)
    except : # 抛出报错行代码、error名
        continue
    
```
